## Results and discussion

The route-screening parameter study compares carbon intensity, energy demand, and maturity across common hydrogen routes. The purpose is to demonstrate how NeqSim cases become decision evidence rather than isolated calculations.

**Calculation basis.** Chapter-specific screening calculations for route screening across grey, blue, green, pink, and selected carrier-adjacent H2 concepts, normalized to the KPIs: route selection, boundary definition, and evidence quality.

**Study basis.** IEA hydrogen route context and the cited process-design references define the interpretation envelope for the normalized indices. \cite{iea2023hydrogen,towler2013chemical}

**Discussion.** The graph shows that route selection is a multi-objective problem. A low-carbon route may need more electrical or thermal energy, and a mature route may not meet long-term carbon targets. The book uses this pattern repeatedly: model the route, expose the assumptions, and discuss the trade-off in engineering terms.
