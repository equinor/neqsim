## Results and discussion

The end-use chapter sweeps hydrogen volume fraction in a methane-dominated gas grid (0, 5, 10, 20, 30, 50 vol%) and tracks Wobbe Index, lower heating value (volumetric), laminar flame speed index, and a NOx-attention index for premix burners.

**Calculation basis.** Chapter-specific screening calculations for the demand side of the hydrogen value chain, normalized to the KPIs: H2 demand by sector, end-use efficiency, blending limits, fuel-quality status, and value-chain matching.

**Study basis.** IEA hydrogen route context and the cited process-design references define the interpretation envelope for the normalized indices. \cite{iea2023hydrogen,towler2013chemical}

**Discussion.** At low blend levels (<10 vol% H2) Wobbe Index stays within typical EN 16726 H-gas bands and most existing appliances tolerate the change. Above 20 vol% the volumetric heating value drops noticeably and flame speed rises sharply, which can shift flame stability windows and NOx formation. End-use compatibility is therefore a per-appliance question, not a single grid-wide tolerance.
